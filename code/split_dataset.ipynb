```python
import pandas as pd
```

```python
file_path = "../dataset/anchor_labeled_data.csv"
df = pd.read_csv(file_path, dtype=str, encoding="cp949")
df.columns = df.columns.str.strip()

if "RISK_LABEL" not in df.columns:
    raise KeyError(f"'RISK_LABEL' column not found. Available columns: {list(df.columns)}")
```

```python
test_df = df.groupby("RISK_LABEL", group_keys=False).sample(frac=0.1, random_state=42).copy()

train_df = df.drop(index=test_df.index).copy()
test_df = test_df.sort_index().reset_index(drop=True)
train_df = train_df.sort_index().reset_index(drop=True)

train_output_path = "../dataset/train.csv"
test_output_path = "../dataset/test.csv"

train_df.to_csv(train_output_path, index=False, encoding="cp949")
test_df.to_csv(test_output_path, index=False, encoding="cp949")

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
print("\nTrain label distribution:")
print(train_df["RISK_LABEL"].value_counts().sort_index())
print("\nTest label distribution:")
print(test_df["RISK_LABEL"].value_counts().sort_index())
```
