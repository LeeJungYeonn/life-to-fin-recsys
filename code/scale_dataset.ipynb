```python
import pandas as pd
```

```python
df = pd.read_csv("../dataset/sub-data.csv", encoding='cp949')
```

```python
# 1. 돈과 관련된 칼럼 리스트 정의
money_cols = [
    'ASSET', 'FIN', 'NFIN', 'LIQ', 'CHECKING', 'SAVING', 'MMA', 'CDS', 
    'STOCKS', 'EQUITY', 'DEQ', 'BOND', 'GOVTBND', 'OBND', 'MORTBND', 
    'RETQLIQ', 'RETEQ', 'IRAKH'
]

# 2. 모든 돈 관련 칼럼에 해당 row의 CPI_DEFL 값을 곱하기 (axis=0을 통해 행 단위 곱셈 수행)
df[money_cols] = df[money_cols].multiply(df['CPI_DEFL'], axis=0)

# 3. 작업이 끝난 후 불필요해진 YEAR, CPI_DEFL 칼럼 삭제
df = df.drop(columns=['YEAR', 'CPI_DEFL'])

# 결과 확인
print(df.head())
```

```python
# 저장
output_path = "../dataset/sub-data.csv"
df.to_csv(output_path, index=False, encoding="cp949")

print(f"비고 행 보존 + 데이터 {len(df)}행 저장됨")
```
